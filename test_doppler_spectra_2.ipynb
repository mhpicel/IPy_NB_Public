```python
#nc = Dataset('/Users/mvanlier/Dropbox/TWP_sband/sgpsbdprecipspcC1.a2.20110520.120000.nc','r')
#nc = Dataset('/home/mvanlier/Dropbox/TWP_sband/sgpsbdprecipspcC1.a2.20110520.120000.nc','r')
nc = Dataset('/home/mvanlier/Dropbox/TWP_sband/sgp1sbdprecipspcC1.a1.20110520.110000.nc','r')
#spec_twp = nc.variables['precip_adj_cal_spc'][:,:,:]
spec_sgp = nc.variables['precip_cal_spc'][:,:,:]
sgp_hour = nc.variables['precip_time_hour'][:]
sgp_minu = nc.variables['precip_time_minute'][:]
sgp_secs = nc.variables['precip_time_second'][:]
sgp_vd   = nc.variables['precip_Vd'][:]
sgp_zax  = nc.variables['precip_range'][:]
sgp_tdbz = nc.variables['precip_zdb'][:,:]
nc.close()
#..
nc = Dataset('/home/mvanlier/Dropbox/TWP_sband/sgp1sbdprecipspcC1.a1.20110520.120000.nc','r')
#spec_twp2 = nc.variables['precip_cal_spc'][:,:,:]
spec_sgp2 = nc.variables['precip_cal_spc'][:,:,:]
sgp_hour2 = nc.variables['precip_time_hour'][:]
sgp_minu2 = nc.variables['precip_time_minute'][:]
sgp_secs2 = nc.variables['precip_time_second'][:]
sgp_vd2   = nc.variables['precip_Vd'][:]
sgp_zax2  = nc.variables['precip_range'][:]
sgp_tdbz2 = nc.variables['precip_zdb'][:,:]
#..
#std_spc  = nc.variables['precip_adj_std_spc'][:,:]            #..(time,range)
#std_mdv  = nc.variables['precip_adj_Vmean_uncertainty'][:,:]  #..(time,range)
#std_dbz  = nc.variables['precip_adj_zdb_uncertainty'][:,:]    #..(time,range)
#..
nc.close()
```

```python
#..Concatenate time variables, convert to Julian time
sgp_time = 20 +np.r_[sgp_hour,sgp_hour2]/24. +np.r_[sgp_minu,sgp_minu2]/(24.*60.) +np.r_[sgp_secs,sgp_secs2]/(24.*3600.)
print sgp_time[0],sgp_time[-1]
```

```python
#..Find indices of desired start and end times
ts_sgp = 20.46
te_sgp = 20.52
tidx_s1  = min(range(len(sgp_time)),key=lambda i:abs(sgp_time[i]-ts_sgp))
tidx_e1  = min(range(len(sgp_time)),key=lambda i:abs(sgp_time[i]-te_sgp))
print tidx_s1, tidx_e1
```

```python
#..Concatenate reflectivity (dbz) profiles together
sgp_tdbz_all = np.concatenate((sgp_tdbz,sgp_tdbz2),axis=0)
print sgp_tdbz_all.shape
```

```python
#..Calculate average reflectivity profile from 'precip_zdb'
sgp_tref_all = 10.**(sgp_tdbz_all/10.)
sgp_tdbz_avg = 10.*np.log10(sgp_tref_all[tidx_s1:tidx_e1,:].mean(axis=0,dtype=np.float64))
```

```python
#..Concatenate spectra
sgp_spc_all = np.concatenate((spec_sgp,spec_sgp2),axis=2)
print sgp_spc_all.shape
print tidx_s1,tidx_e1
```

```python
#..Mask out unphysical values
sgp_spc_all = np.ma.masked_where(sgp_spc_all<0.,sgp_spc_all)
```

```python
#..Find the average spectrum
sgp_spc_avg = sgp_spc_all[:,:,tidx_s1:tidx_e1+1].mean(axis=2,dtype=np.float64)
sgp_spc_std = np.std(sgp_spc_all[:,:,tidx_s1:tidx_e1+1],axis=2)
```

```python
#..Calculate spectrum in dB, plot
sgp_spc_avg_c = copy.deepcopy(sgp_spc_avg)
sgp_spc_avg_c = np.ma.masked_where(sgp_spc_avg==0.,sgp_spc_avg)
sgp_logspec = 10.*np.log10(sgp_spc_avg_c)
plt.pcolormesh(sgp_vd,sgp_zax,sgp_logspec.T,vmin=-30,vmax=40,cmap=cm.gray)
plt.colorbar()
```

```python
#..Calculate average reflectivity in dBZ for average spectrum
sgp_ref = sgp_spc_std.sum(axis=0)*(sgp_vd[1]-sgp_vd[0])
sgp_dbz = 10.*np.log10(sgp_ref)
```

```python
#..Compare average relectivity profile in dBZ from spectrum (blue) and from dBZ profiles (green)
fig = plt.figure(figsize=(5,10))
ax1 = fig.add_subplot(111)
plt.plot(sgp_dbz,sgp_zax,label='dBZ from spectrum')
plt.plot(sgp_tdbz_avg,sgp_zax,label='dBZ from file')
plt.title('Average Reflectivity (dBZ)')
ax1.legend()
ax1.set_ylim(0,12000)
ax1.set_xlim(0,50)
```
