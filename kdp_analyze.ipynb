The next section details analysis of strong kdp "tower" features. 

```python
kdpi[kdpi < 2.5] =0.
kdpi = kdpi
kdpif = ndi.gaussian_filter(kdpi,1)
fig = plt.figure(figsize=(13,10))
plt.contourf(xd_p,xd_p,kdpif,40)
plt.colorbar()
```

```python
#kdpif = kdpif.max() - kdpif
#kdpif -= kdpif.min()
kdpif = kdpif/float(kdpif.ptp())*255
print kdpif.min()
print kdpif.max()
kdpif = kdpif.astype(np.uint8)
rmax  = pymorph.regmax(kdpif)
fig = plt.figure(figsize=(14,10))
pylab.imshow(pymorph.overlay(kdpif,rmax),origin='lower')
#pylab.show()
seeds,nr_nuclei = ndi.label(rmax)
print 'number of nuclei = ',nr_nuclei
```

```python
T =mahotas.thresholding.otsu(kdpif)
print kdpif.ptp()
print T
dist = ndi.distance_transform_edt(kdpif > T)
fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot(1,2,1)
plt.imshow(dist,origin='lower')
dist = dist.max() - dist
dist -= dist.min()
dist = dist/float(dist.ptp())*255
dist = dist.astype(np.uint8)
nuclei = pymorph.cwatershed(dist,seeds)
ax1 = fig.add_subplot(1,2,2)
plt.imshow(nuclei,origin='lower')
#plt.contourf(lon_p,lat_p,nuclei,25,cmap=cm.prism)
plt.colorbar()
```

```python
xextent = np.ceil(xextent/2)
yextent = np.ceil(yextent/2)
zhh[zhh<0.]=0.
kdp_cut = kdp[:,yidxmin-yextent:yidxmax+yextent,xidxmin-xextent:xidxmax+xextent]
xcut = xd_p[xidxmin-xextent:xidxmax+xextent]
ycut = yd_p[yidxmin-yextent:yidxmax+yextent]
print kdp_cut.shape
print xcut.shape
print ycut.shape
print zd_p.shape
print zd_p.ptp()
print xcut.ptp()
print ycut.ptp()
xcut -= xcut.min()
ycut -= ycut.min()
xg,yg,zg = np.meshgrid(ycut,xcut,zd_p,indexing='xy')
#xg,yg,zg = np.meshgrid(xcut,ycut,zd_p,indexing='xy')
#xg,yg,zg = np.meshgrid(xcut,ycut,zd_p,indexing='ij')
print xg.shape
plt.pcolormesh(xg[:,0,:])
kdp_p = np.transpose(kdp_cut,(2,1,0))
#kdp_p = np.transpose(kdp_cut,(1,2,0))
print kdp_p.shape
```

```python
tst = mlab.pipeline.volume(mlab.pipeline.scalar_field(yg,xg,zg,kdp_p),vmin=1.5,vmax=2.5)
#mlab.outline()
#mlab.colorbar()
mlab.axes(x_axis_visibility=True,y_axis_visibility=True,z_axis_visibility=True)
mlab.show()
```

```python
#..Set parameters for plotting
kdp_max   = 2.5 #..Max isosurface value
kdp_min   = 1.  #..Min isosurface value
num_iso   = 4   #..Number of isosurfaces
def_alph  = .5  #..Default alpha for isosurfaces
max_idx = 241

fig = plt.figure(figsize=(16,10))
for j in range(1,nr_nuclei):
    #..First, figure out extent of region identified by nucleus number i
    xgrid_cut = xgrid[nuclei == j]
    ygrid_cut = ygrid[nuclei == j]
    print xgrid_cut.min(),xgrid_cut.max(),ygrid_cut.min(),ygrid_cut.max()
    xidxmax2= np.argmin(np.abs(xd_p - xgrid_cut.max()))
    xidxmin2= np.argmin(np.abs(xd_p - xgrid_cut.min()))
    yidxmax2= np.argmin(np.abs(yd_p - ygrid_cut.max()))    
    yidxmin2= np.argmin(np.abs(yd_p - ygrid_cut.min()))
    xidxmax = min(range(len(xd_p)),key=lambda i:abs(xd_p[i]-xgrid_cut.max()))
    xidxmin = min(range(len(xd_p)),key=lambda i:abs(xd_p[i]-xgrid_cut.min()))
    yidxmax = min(range(len(yd_p)),key=lambda i:abs(yd_p[i]-ygrid_cut.max()))
    yidxmin = min(range(len(yd_p)),key=lambda i:abs(yd_p[i]-ygrid_cut.min()))
    xextent = abs(xidxmax-xidxmin)
    yextent = abs(yidxmax-yidxmin)
    print xidxmin,xidxmax,yidxmin,yidxmax
    print xidxmin2,xidxmax2,yidxmin2,yidxmax2
    
    print 'xextent=',xextent,'yextent=',yextent
    #..Now make cartesian grid for these regions (with a boundary ~ extent)
    xextent = max(np.ceil(xextent/2),4)
    yextent = max(np.ceil(yextent/2),4)
    kdp_cut = kdp[:,max(0,yidxmin-yextent):min(max_idx,yidxmax+yextent),
                    max(0,xidxmin-xextent):min(max_idx,xidxmax+xextent)]
    xcut = xd_p[max(0,xidxmin-xextent):min(max_idx,xidxmax+xextent)]
    ycut = yd_p[max(0,yidxmin-yextent):min(max_idx,yidxmax+yextent)]
    print type(xd_p)
    print type(yd_p)
    print xcut.shape
    print ycut.shape
    if len(xcut)!=1:
        xcut -= xcut.min()
    if len(ycut)!=1:
        ycut -= ycut.min()
    xg,yg,zg = np.meshgrid(ycut,xcut,zd_p,indexing='xy')
    #..Rearrange kdp_cut into the shape that mayavi expects
    kdp_p = np.transpose(kdp_cut,(2,1,0))

    #..Display mayavi mlab object
    contours = list(np.linspace(kdp_min,kdp_max,num_iso))
    tst = mlab.contour3d(yg,xg,zg,kdp_p,contours=contours,opacity=def_alph)
    mfig = mlab.gcf()
    print type(tst)
    mlab.axes(x_axis_visibility=True,y_axis_visibility=True,z_axis_visibility=True)
    marr = mlab.screenshot(mfig)
    mlab.close(mfig)
    #mlab.close(all=True)
    ax1 = fig.add_subplot(2,np.ceil(nr_nuclei/2),nr_nuclei)
    plt.imshow(marr)
```
