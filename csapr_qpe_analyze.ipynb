```python
def area_anal(pixel_area, rr_x, rain_rates):
    A_rainrate=np.zeros(rr_x.shape)
    N_rainrate=np.zeros(rr_x.shape)
    Rm_rainrate=np.zeros(rr_x.shape)
    my_shape=rain_rates.shape
    for i in range(len(rr_x)):
        b_fld=np.zeros(my_shape)
        b_fld[rain_rates > rr_x[i]]=1.0
        regions, N_rainrate[i] = ndimage.label(b_fld)
        try:
            A_rainrate[i]=len(np.where(regions > 0.5)[0])*pixel_area
            Rm_rainrate[i]=rain_rates[np.where(regions > 0.5)].mean()
        except IndexError:
            A_rainrate[i]=0.0
            Rm_rainrate[i]=0.0
    return N_rainrate, A_rainrate, Rm_rainrate
```

```python
def plot_qpe_ts(path,daystring,ts=None,te=None):
    qpefiles = []
    numfiles = 0
    origdir = os.getcwd()
    os.chdir(path)
    for root,dirs,files in os.walk('./'):
        for file in files:
            if file.startswith('sgpcsaprqpe'):
                qpefiles.append(file)
                numfiles += 1
    nx=241
    ny=241
    rr_all = np.empty([numfiles,ny,nx])
    tm_all = np.empty([numfiles])
    for i in range(numfiles):
        filenm = qpefiles[i]
        nc = Dataset(filenm,'r')
        rr_all[i,:,:] = nc.variables['rain_rate_A'][0,0,:,:]
        tm_all[i] = ((datetime.datetime(int(filenm[17:21]),int(filenm[21:23]),int(filenm[23:25]),
                                        int(filenm[26:28]),int(filenm[28:30]),int(filenm[30:32])) -
                      datetime.datetime(1970,1,1)).total_seconds())
        nc.close()
    os.chdir(origdir)
    bin_width = 2.
    rmin = 0.
    rmax = 130.
    rbins = np.arange(rmin,rmax,bin_width)
    rcent = np.arange(rmin+bin_width/2,rmax-bin_width/2,bin_width)
    his_t = np.empty([numfiles,rbins.shape[0]-1])
    for i in range(numfiles):
        dum,edge = np.histogram(rr_all[i,:,:].flatten(),rbins,weights=rr_all[i,:,:].flatten())
        #print dum.shape
        his_t[i,:] = dum
    if not(ts):
        ts=0
    if not(te):
        te=numfiles
    print numfiles
    hfmt = dates.DateFormatter('%H:%M')
    dts  = map(datetime.datetime.utcfromtimestamp,tm_all)
    tim  = dates.date2num(dts)
    f=plt.figure(figsize=[15,9])
    plt.pcolormesh(tim[ts:te],rcent,his_t[ts:te,:].T,cmap=cm.spectral_r,vmin=0,vmax=24000.)
    ax1 = plt.gca()
    ax1.xaxis.set_major_locator(dates.HourLocator())
    ax1.xaxis.set_major_formatter(hfmt)
    plt.title('Weighted rainfall histogram with time, '+daystring)
    plt.autoscale(enable=True, axis='both', tight=True)
    plt.colorbar()
```
